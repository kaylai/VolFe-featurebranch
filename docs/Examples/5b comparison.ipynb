## Decide appropriate models

```python
CO2_calibration = pd.read_csv("../files/melt_chemistry_CO2_solubility.csv")  # load data
```

```python
def sum(comp):
    sum = comp['SiO2']+comp['TiO2']+comp['Al2O3']+comp['FeOT']+comp['MnO']+comp['MgO']+comp['CaO']+comp['Na2O']+comp['K2O']+comp['P2O5']
    return sum
```

```python
sum_Kilauea_dry = sum(comp_Kilauea_dry)
sum_MORB = sum(comp_MORB)
sum_Fogo = sum(comp_Fogo)
sum_Fuego = sum(comp_Fuego)
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, figsize=(16,12))

oxides = ["TiO2",'Al2O3','FeOT','MgO','MnO','CaO','Na2O','K2O','P2O5']
exp_symbols = pd.DataFrame(['*','o','s','D','X','^','v','<','>','P','h','H','d','p'])

n=0
# Plotting results
for x in oxides:
    n = n+1
    m = str(n)
    axi = eval('ax'+m)
    for m in range(0,14):
        axi.plot(100*CO2_calibration.loc[m,'SiO2']/CO2_calibration.loc[m,'total'], 100*CO2_calibration.loc[m,x]/CO2_calibration.loc[m,'total'], marker = str(exp_symbols.iloc[m,0]),mec='black', mfc='grey',markersize=8)
    axi.plot(100*comp_Fuego['SiO2']/sum_Fuego, 100*comp_Fuego[x]/sum_Fuego, 'sk', mfc='darkorange',markersize=10)
    axi.plot(100*comp_Fogo['SiO2']/sum_Fogo, 100*comp_Fogo[x]/sum_Fogo, 'sk', mfc='firebrick',markersize=10)
    axi.plot(100*comp_Kilauea_dry['SiO2']/sum_Kilauea_dry, 100*comp_Kilauea_dry[x]/sum_Kilauea_dry, 'sk', mfc='steelblue',markersize=10)
    axi.plot(100*comp_MORB['SiO2']/sum_MORB, 100*comp_MORB[x]/sum_MORB, 'sk', mfc='darkviolet',markersize=10)
    axi.set_ylabel(x+' (wt%)')
    if axi == ax7 or axi == ax8 or axi == ax9:
        axi.set_xlabel("SiO2 (wt%)")
    axi.set_xlim([42,55])
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9),(ax10,ax11,ax12)) = plt.subplots(4, 3, figsize=(16,16))

oxides = ["TiO2",'Al2O3','FeOT','MgO','MnO','CaO','Na2O','K2O','P2O5','SiO2']
exp_symbols = pd.DataFrame(['*','o','s','D','X','^','v','<','>','P','h','H','d','p'])
comps = [comp_MORB,comp_Kilauea_dry,comp_Fuego,comp_Fogo]

n=0
# Plotting results
for x in oxides:
    n = n+1
    m = str(n)
    axi = eval('ax'+m)
    for c in comps:
        if c.loc[0,'Sample'] == 'MORB':
            col = 'darkviolet'
            num = 2
            sum = sum_MORB
        if c.loc[0,'Sample'] == 'Kīlauea_Dryend':
            col = 'steelblue'
            num = 3
            sum = sum_Kilauea_dry
        if c.loc[0,'Sample'] == 'Fuego':
            col = 'darkorange'
            num = 4
            sum = sum_Fuego
        if c.loc[0,'Sample'] == 'Fogo':
            col = 'firebrick'
            num = 1
            sum = sum_Fogo
        for m in range(0,14):
            diff = (100*c.loc[0,x]/sum)-(100*CO2_calibration.loc[m,x]/CO2_calibration.loc[m,'total'])
            axi.plot(num, diff, marker = str(exp_symbols.iloc[m,0]),mec='black', mfc=col,markersize=8)
    axi.set_ylabel(x+' (wt%)')
    if axi == ax7 or axi == ax8 or axi == ax9:
        axi.set_xlabel("SiO2 (wt%)")
    axi.set_xlim([0,5])
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(7,5))

oxides = ["TiO2",'Al2O3','FeOT','MgO','MnO','CaO','Na2O','K2O','P2O5']
exp_symbols = pd.DataFrame(['*','o','s','D','X','^','v','<','>','P','h','H','d','p'])

for m in range(0,14):
    ax1.plot(100*CO2_calibration.loc[m,'SiO2']/CO2_calibration.loc[m,'total'], (100*(CO2_calibration.loc[m,'Na2O']+CO2_calibration.loc[m,'K2O'])/CO2_calibration.loc[m,'total']), marker = str(exp_symbols.iloc[m,0]),mec='black', mfc='grey',markersize=8)
ax1.plot(100*comp_Fuego['SiO2']/sum_Fuego, (100*(comp_Fuego["Na2O"]+comp_Fuego["K2O"])/sum_Fuego), 'sk', mfc='darkorange',markersize=10)
ax1.plot(100*comp_Fogo['SiO2']/sum_Fogo, (100*(comp_Fogo["Na2O"]+comp_Fogo['K2O'])/sum_Fogo), 'sk', mfc='firebrick',markersize=10)
ax1.plot(100*comp_Kilauea_dry['SiO2']/sum_Kilauea_dry, (100*(comp_Kilauea_dry["Na2O"]+comp_Kilauea_dry["K2O"])/sum_Kilauea_dry), 'sk', mfc='steelblue',markersize=10)
ax1.plot(100*comp_MORB['SiO2']/sum_MORB, (100*(comp_MORB["Na2O"]+comp_MORB["K2O"])/sum_MORB), 'sk', mfc='darkviolet',markersize=10)
```

```python
# MORB = default models (carbon dioxide = MORB_Dixon95)
# Kilauea = default models (carbon dioxide = MORB_Dixon95)

# Fuego
# Choose the options I want to change for the calculation - everything else will use the default options
my_models = [['carbon dioxide','SFVFBasalticAndesite_Allison19']]

# turn to dataframe with correct column headers and indexes    
models_Fuego = vf.make_df_and_add_model_defaults(my_models)


# Fogo
# Choose the options I want to change for the calculation - everything else will use the default options
my_models = [['carbon dioxide','Leucitite_Thibault94']]

# turn to dataframe with correct column headers and indexes    
models_Fogo = vf.make_df_and_add_model_defaults(my_models)
```
